### Step 2: Examine the Structure of the Data

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
import pandas as pd

url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(url)

print("MISSING VALUES (COUNT):")
missing_count = df.isnull().sum().sort_values(ascending=False)
print(missing_count[missing_count > 0])

print("\nMISSING VALUES (%):")
missing_percent = (df.isnull().sum() / len(df)) * 100
print(missing_percent[missing_percent > 0].sort_values(ascending=False))

df_cleaned = df.copy()

for col in df_cleaned.columns:

    if df_cleaned[col].dtype in ['int64', 'float64']:
        df_cleaned[col] = df_cleaned[col].fillna(df_cleaned[col].median())

    else:
        if df_cleaned[col].isnull().sum() > 0:
            df_cleaned[col] = df_cleaned[col].fillna(df_cleaned[col].mode()[0])

print("\nTOTAL MISSING VALUES AFTER CLEANING:")
print(df_cleaned.isnull().sum().sum())


df_cleaned.to_csv("cleaned_survey_data.csv", index=False)

print("\nCleaned dataset saved as 'cleaned_survey_data.csv'")
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(url)
jobsat = df['JobSat'].dropna()

counts = jobsat.value_counts()

plt.figure(figsize=(8, 8))

plt.pie(
    counts,
    labels=counts.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.viridis.colors
)

plt.title("Distribution of Job Satisfaction (JobSat)")
plt.axis('equal')
plt.show()
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(url)
lang_df = df[['LanguageHaveWorkedWith', 'LanguageWantToWorkWith']].dropna()

lang_df['Have'] = lang_df['LanguageHaveWorkedWith'].str.split(';')
lang_df['Want'] = lang_df['LanguageWantToWorkWith'].str.split(';')

have = lang_df.explode('Have')
want = lang_df.explode('Want')

have['Have'] = have['Have'].str.strip()
want['Want'] = want['Want'].str.strip()

have_counts = have['Have'].value_counts()
want_counts = want['Want'].value_counts()

compare = pd.DataFrame({
    'HaveWorked': have_counts,
    'WantToWork': want_counts
}).fillna(0)


top_langs = (have_counts + want_counts).sort_values(ascending=False).head(10).index
compare = compare.loc[top_langs]

compare.plot(kind='bar', figsize=(12,6), colormap='viridis')

plt.title("Have Worked vs Want to Work Languages")
plt.xlabel("Programming Language")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(url)
data = df[['Country', 'RemoteWork']].dropna()

remote_ct = pd.crosstab(data['Country'], data['RemoteWork'])

top_countries = data['Country'].value_counts().head(10).index
remote_ct = remote_ct.loc[top_countries]

remote_ct.plot(
    kind='bar',
    figsize=(14, 6),
    colormap='viridis'
)

plt.title("Remote Work Distribution by Region (Top 10 Countries)")
plt.xlabel("Country")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=45, ha='right')
plt.legend(title="Remote Work Status")
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))

sns.heatmap(
    remote_ct,
    cmap="YlGnBu",
    linewidths=0.5
)

plt.title("Remote Work Distribution by Region (Heatmap)")
plt.xlabel("Remote Work Status")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
import pandas as pd

url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(url)

data = df[['JobSat', 'YearsCodePro']].copy()

data['YearsCodePro'] = pd.to_numeric(data['YearsCodePro'], errors='coerce')

data = data.dropna()

jobsat_map = {
    'Very dissatisfied': 1,
    'Dissatisfied': 2,
    'Neither satisfied nor dissatisfied': 3,
    'Satisfied': 4,
    'Very satisfied': 5
}

data['JobSatScore'] = data['JobSat'].map(jobsat_map)

data = data.dropna()

pearson_corr = data['YearsCodePro'].corr(data['JobSatScore'], method='pearson')

spearman_corr = data['YearsCodePro'].corr(data['JobSatScore'], method='spearman')

print("Correlation Analysis:")
print(f"Pearson Correlation : {pearson_corr:.3f}")
print(f"Spearman Correlation: {spearman_corr:.3f}")
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(url)

data = df[['Employment', 'EdLevel']].dropna()

emp_ed_pct = pd.crosstab(
    data['EdLevel'],
    data['Employment'],
    normalize='index'
) * 100

ax = emp_ed_pct.plot(
    kind='bar',
    stacked=True,
    figsize=(16, 8),
    colormap='viridis'
)

plt.title("Employment Status by Education Level")
plt.xlabel("Education Level")
plt.ylabel("Percentage of Respondents")

plt.xticks(rotation=45, ha='right')

plt.legend(
    title="Employment Type",
    bbox_to_anchor=(1.05, 1),
    loc='upper left'
)

plt.subplots_adjust(left=0.25, right=0.75, bottom=0.3, top=0.9)

plt.show()
```
