### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python

import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter


df = pd.read_csv("survey-data.csv")

database_counter = Counter()

for databases in df["DatabaseWantToWorkWith"].dropna():
    for db in databases.split(";"):
        database_counter[db.strip()] += 1

top5 = database_counter.most_common(5)

labels = [item[0] for item in top5]
sizes = [item[1] for item in top5]

plt.figure(figsize=(8, 8))

plt.pie(
    sizes,
    labels=labels,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Top 5 Databases Respondents Want to Work With")

plt.tight_layout()
plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.

##### 3.1 Pie Chart of `Respondents` Most Admired Programming Languages


The `LanguageAdmired` column lists the programming languages respondents admire most. Create a pie chart to visualize the top 5 admired languages.



```python
%pip install pandas matplotlib -q

from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("survey-data.csv")

lang_counter = Counter()

for response in df["LanguageAdmired"].dropna():
    for lang in response.split(";"):
        lang_counter[lang.strip()] += 1

top_5 = lang_counter.most_common(5)

labels = [lang for lang, count in top_5]
sizes = [count for lang, count in top_5]

plt.figure(figsize=(8, 8))

plt.pie(
    sizes,
    labels=labels,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Top 5 Most Admired Programming Languages")

plt.tight_layout()
plt.show()
```

##### 3.2 Pie Chart of Tools Used for AI Development


Using the `AIToolCurrently` Using column, create a pie chart to visualize the top 5 tools developers are currently using for AI development.



```python
%pip install pandas matplotlib -q

from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("survey-data.csv")

ai_col = None
for col in df.columns:
    if "AI" in col and "Tool" in col or "AIDev" in col or "AISelect" in col:
        ai_col = col
        break

print("Using column:", ai_col)

tool_counter = Counter()

for response in df[ai_col].dropna():
    tool_counter.update([tool.strip() for tool in str(response).split(";")])

top_5 = tool_counter.most_common(5)

labels = [tool for tool, count in top_5]
sizes = [count for tool, count in top_5]

plt.figure(figsize=(8, 8))

plt.pie(
    sizes,
    labels=labels,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Top 5 AI Development Tools Currently Used")

plt.tight_layout()
plt.show()
```
