### Task 2: Analysis of Duplicate Response Patterns
1. Identify duplicate response patterns based on selected columns such as MainBranch, Employment, and RemoteWork.
2. Clarify that these represent multiple respondents with identical answers rather than duplicate records. Analyse which other columns frequently share identical values within these response-pattern groups.
   


```python
import pandas as pd

url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/VYPrOu0Vs3I0hKLLjiPGrA/survey-data-with-duplicate.csv"
df = pd.read_csv(url)

pattern_cols = ['MainBranch', 'Employment', 'RemoteWork']

pattern_groups = df.groupby(pattern_cols)

duplicate_patterns = pattern_groups.filter(lambda x: len(x) > 1)

print("Rows sharing identical response patterns:")
print(duplicate_patterns.head())

print("\nNumber of rows in repeated response-pattern groups:")
print(len(duplicate_patterns))

print("\nMost common response patterns:")
print(
    df.groupby(pattern_cols)
      .size()
      .reset_index(name='Count')
      .sort_values(by='Count', ascending=False)
      .head(10)
)

other_columns = [col for col in df.columns if col not in pattern_cols]

matching_summary = {}

for col in other_columns:
  
    consistency = pattern_groups[col].nunique()
    

    identical_groups = (consistency == 1).sum()
    
    matching_summary[col] = identical_groups

matching_df = pd.DataFrame(
    matching_summary.items(),
    columns=['Column', 'Groups_with_Identical_Values']
)

matching_df = matching_df.sort_values(
    by='Groups_with_Identical_Values',
    ascending=False
)

print("\nColumns frequently sharing identical values:")
print(matching_df.head(10))
```

### Task 3: Visualize Shared Response Patterns
1. Create visualizations to show the distribution of shared response patterns across different categories.
2. Use bar charts or pie charts to represent the distribution of respondents who share identical values for MainBranch, Employment, and RemoteWork, grouped by Country and Employment.


```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/VYPrOu0Vs3I0hKLLjiPGrA/survey-data-with-duplicate.csv"

df = pd.read_csv(url)

pattern_cols = ['MainBranch', 'Employment', 'RemoteWork']

pattern_counts = (
    df.groupby(pattern_cols)
      .size()
      .reset_index(name='RespondentCount')
)

shared_patterns = pattern_counts[
    pattern_counts['RespondentCount'] > 1
]

merged_df = df.merge(
    shared_patterns[pattern_cols],
    on=pattern_cols,
    how='inner'
)

country_counts = (
    merged_df['Country']
      .value_counts()
      .head(10)
)

plt.figure(figsize=(14,7))

plt.bar(
    country_counts.index,
    country_counts.values,
    color='steelblue'
)

plt.title('Top Countries with Shared Response Patterns')
plt.xlabel('Country')
plt.ylabel('Number of Respondents')

plt.xticks(rotation=45)

# Better spacing
plt.subplots_adjust(bottom=0.3)

plt.show()

employment_counts = merged_df['Employment'].value_counts()

plt.figure(figsize=(14,7))

plt.bar(
    employment_counts.index,
    employment_counts.values,
    color='darkorange'
)

plt.title('Employment Distribution in Shared Patterns')
plt.xlabel('Employment Type')
plt.ylabel('Number of Respondents')

plt.xticks(rotation=45)

# Better spacing
plt.subplots_adjust(bottom=0.4)

plt.show()

top_employment = employment_counts.head(5)

plt.figure(figsize=(8,8))

plt.pie(
    top_employment.values,
    labels=top_employment.index,
    autopct='%1.1f%%',
    startangle=140
)

plt.title('Employment Share within Shared Response Patterns')

plt.show()
```
