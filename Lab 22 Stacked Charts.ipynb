### Task 1: Stacked Chart for Composition of Job Satisfaction Across Age Groups


##### 1. Stacked Chart of Median `JobSatPoints_6` and `JobSatPoints_7` for Different Age Groups


Visualize the composition of job satisfaction scores (`JobSatPoints_6` and `JobSatPoints_7`) across various age groups. This will help in understanding the breakdown of satisfaction levels across different demographics.



```python
%pip install pandas matplotlib -q

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("survey-data.csv")

median_scores = (
    df.groupby("Age")[["JobSatPoints_6", "JobSatPoints_7"]]
      .median()
      .dropna()
)

ax = median_scores.plot(
    kind="bar",
    stacked=True,
    figsize=(12, 6)
)

plt.title("Median Job Satisfaction Scores by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Median Satisfaction Score")
plt.legend(["JobSatPoints_6", "JobSatPoints_7"])
plt.tight_layout()
plt.show()
```

##### Stacked Chart of `JobSatPoints_6` and `JobSatPoints_7` for Employment Status


Create a stacked chart to compare job satisfaction (`JobSatPoints_6` and `JobSatPoints_7`) across different employment statuses. This will show how satisfaction varies by employment type.


```python
%pip install pandas matplotlib -q

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("survey-data.csv")

jobsat_by_employment = (
    df.groupby("Employment")[["JobSatPoints_6", "JobSatPoints_7"]]
      .median()
      .dropna()
)
ax = jobsat_by_employment.plot(
    kind="bar",
    stacked=True,
    figsize=(12, 6)
)

plt.title("Job Satisfaction by Employment Status")
plt.xlabel("Employment Status")
plt.ylabel("Median Job Satisfaction Score")

plt.legend(["JobSatPoints_6", "JobSatPoints_7"])

plt.xticks(rotation=45, ha="right")

plt.subplots_adjust(bottom=0.25)

plt.show()
```

### Task 2: Stacked Chart for Compensation and Job Satisfaction by Age Group


##### This stacked chart visualizes the composition of compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`) specifically for respondents aged 30-35.


```python
%pip install pandas matplotlib -q

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("survey-data.csv")

age_group = df[df["Age"] == "25-34 years old"]

median_comp = age_group["ConvertedCompYearly"].median()
median_jobsat = age_group["JobSatPoints_6"].median()

stacked_data = pd.DataFrame({
    "ConvertedCompYearly": [median_comp],
    "JobSatPoints_6": [median_jobsat]
}, index=["Age 25-34"])

ax = stacked_data.plot(
    kind="bar",
    stacked=True,
    figsize=(8, 6)
)
plt.title("Median Compensation and Job Satisfaction (Age 25-34)")
plt.ylabel("Median Value")
plt.legend(["ConvertedCompYearly", "JobSatPoints_6"])
plt.xticks(rotation=0)

plt.tight_layout()
plt.show()
```

##### Stacked Chart of Median Compensation and Job Satisfaction Across Age Group


Compare the median compensation and job satisfaction metrics across different age groups. This helps visualize how compensation and satisfaction levels differ by age.


```python

%pip install pandas matplotlib scikit-learn -q

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv("survey-data.csv")

age_summary = (
    df.groupby("Age")[["ConvertedCompYearly", "JobSatPoints_6"]]
      .median()
      .dropna()
)

scaler = MinMaxScaler()

normalized = pd.DataFrame(
    scaler.fit_transform(age_summary),
    columns=age_summary.columns,
    index=age_summary.index
)

ax = normalized.plot(
    kind="bar",
    stacked=True,
    figsize=(12, 6)
)

plt.title("Normalized Median Compensation and Job Satisfaction Across Age Groups")
plt.xlabel("Age Group")
plt.ylabel("Normalized Value (0–1)")
plt.legend(["ConvertedCompYearly", "JobSatPoints_6"])
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
```

### Task 3: Comparing Data Using Stacked Charts


##### 1. Stacked Chart of Preferred Databases by Age Group




Visualize the top databases that respondents from different age groups wish to learn. Create a stacked chart to show the proportion of each database in each age group.


```python

%pip install pandas matplotlib -q

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("survey-data.csv")

print(df.columns)

data = df[['Age', 'DatabaseWantToWorkWith']].dropna()

rows = []

for _, row in data.iterrows():
    age = row['Age']
    databases = str(row['DatabaseWantToWorkWith']).split(';')
    for db in databases:
        rows.append([age, db.strip()])

long_df = pd.DataFrame(rows, columns=['Age', 'Database'])

age_db_counts = pd.crosstab(long_df['Age'], long_df['Database'])

age_db_props = age_db_counts.div(age_db_counts.sum(axis=1), axis=0)

top5 = long_df['Database'].value_counts().head(5).index

age_db_props[top5].plot(
    kind='bar',
    stacked=True,
    figsize=(12,6)
)
plt.title('Preferred Databases by Age Group (Top 5)')
plt.xlabel('Age Group')
plt.ylabel('Proportion')
plt.legend(title='Database', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

##### 2. Stacked Chart of Employment Type by Job Satisfaction


Analyze the distribution of employment types within each job satisfaction level using a stacked chart. This will provide insights into how employment types are distributed across various satisfaction ratings.


```python
%pip install pandas matplotlib -q

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("survey-data.csv")

data = df[["Employment", "JobSatPoints_6"]].dropna()

emp_jobsat = pd.crosstab(data["JobSatPoints_6"], data["Employment"])

emp_jobsat_props = emp_jobsat.div(emp_jobsat.sum(axis=1), axis=0)

fig, ax = plt.subplots(figsize=(18, 8))

emp_jobsat_props.plot(
    kind="bar",
    stacked=True,
    ax=ax
)

ax.set_title("Employment Type Distribution by Job Satisfaction")
ax.set_xlabel("Job Satisfaction (JobSatPoints_6)")
ax.set_ylabel("Proportion")

ax.legend(
    title="Employment Type",
    bbox_to_anchor=(1.02, 1),
    loc="upper left",
    fontsize=8
)

plt.xticks(rotation=0)

plt.subplots_adjust(right=0.70)
```

### Task 4: Exploring Technology Preferences Using Stacked Charts


##### 1. Stacked Chart for Preferred Programming Languages by Age Group


Analyze how programming language preferences (`LanguageAdmired`) vary across age groups.


```python
%pip install pandas matplotlib -q

import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("survey-data.csv")

data = df[["Age", "LanguageAdmired"]].dropna()

rows = []

for _, row in data.iterrows():
    age = row["Age"]
    for lang in str(row["LanguageAdmired"]).split(";"):
        rows.append([age, lang.strip()])

long_df = pd.DataFrame(rows, columns=["Age", "Language"])

age_lang = pd.crosstab(long_df["Age"], long_df["Language"])

age_lang = age_lang.div(age_lang.sum(axis=1), axis=0)

top5 = long_df["Language"].value_counts().head(5).index

fig, ax = plt.subplots(figsize=(14, 7))

age_lang[top5].plot(
    kind="bar",
    stacked=True,
    ax=ax
)

ax.set_title("Preferred Programming Languages (Admired) by Age Group")
ax.set_xlabel("Age Group")
ax.set_ylabel("Proportion")
ax.legend(title="Language", bbox_to_anchor=(1.02, 1), loc="upper left")

plt.xticks(rotation=45, ha="right")
plt.subplots_adjust(right=0.75)

plt.show()
```

##### 2. Stacked Chart for Technology Adoption by Employment Type


Explore how admired platforms (`PlatformAdmired`) differ across employment types (e.g., full-time, freelance)


```python
%pip install pandas matplotlib -q

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("survey-data.csv")

print(df.columns)

data = df[["Employment", "PlatformAdmired"]].dropna()

rows = []

for _, row in data.iterrows():
    employment = row["Employment"]
    platforms = str(row["PlatformAdmired"]).split(";")

    for platform in platforms:
        rows.append([employment, platform.strip()])

long_df = pd.DataFrame(rows, columns=["Employment", "Platform"])

emp_platform = pd.crosstab(long_df["Employment"], long_df["Platform"])

emp_platform = emp_platform.div(emp_platform.sum(axis=1), axis=0)

top5 = long_df["Platform"].value_counts().head(5).index

fig, ax = plt.subplots(figsize=(16, 8))

emp_platform[top5].plot(
    kind="bar",
    stacked=True,
    ax=ax
)

ax.set_title("Admired Platforms Across Employment Types (Top 5)")
ax.set_xlabel("Employment Type")
ax.set_ylabel("Proportion")
ax.legend(title="Platform", bbox_to_anchor=(1.02, 1), loc="upper left")

plt.xticks(rotation=45, ha="right")
plt.subplots_adjust(right=0.75)

plt.show()
```
