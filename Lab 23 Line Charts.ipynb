### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
%pip install pandas matplotlib -q

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("survey-data.csv")

age_comp = (
    df.groupby("Age")["ConvertedCompYearly"]
      .median()
      .dropna()
      .sort_index()
)

plt.figure(figsize=(10, 5))
plt.plot(age_comp.index, age_comp.values, marker="o")

plt.title("Median Yearly Compensation by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Median ConvertedCompYearly")
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python
%pip install pandas matplotlib -q

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("survey-data.csv")

df = df[["Age", "ConvertedCompYearly"]].dropna()

age_filtered = df[df["Age"].isin([
    "25-34 years old",
    "35-44 years old"
])]

age_comp = (
    age_filtered.groupby("Age")["ConvertedCompYearly"]
    .median()
    .sort_index()
)

plt.figure(figsize=(8, 5))
plt.plot(age_comp.index, age_comp.values, marker="o")

plt.title("Median ConvertedCompYearly (Ages 25–45)")
plt.xlabel("Age Group")
plt.ylabel("Median Compensation")
plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
```

### Task 2: Trends in Job Satisfaction by Experience Level



##### 1. Line Chart of Job Satisfaction (`JobSatPoints_6`) by Experience Level



- Use a column that approximates experience level to analyze how job satisfaction changes with experience.

- If needed, substitute an available experience-related column for `Experience`.


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("survey-data.csv")

data = df[['YearsCodePro', 'JobSatPoints_6']].dropna()

data = data[data['YearsCodePro'].str.contains(r'\d', na=False)]

data['YearsCodePro'] = data['YearsCodePro'].str.extract('(\d+)').astype(float)

exp_sat = data.groupby('YearsCodePro')['JobSatPoints_6'].median().sort_index()

plt.figure(figsize=(10, 5))
plt.plot(exp_sat.index, exp_sat.values, marker='o')

plt.title('Job Satisfaction by Experience Level')
plt.xlabel('Years of Experience')
plt.ylabel('Median Job Satisfaction')
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

### Task 3: Trends in Job Satisfaction and Compensation by Experience


##### 1.Line Chart of Median ConvertedCompYearly Over Experience Level

- This line chart will track how median compensation (`ConvertedCompYearly`) changes with increasing experience.

- Use a column such as `WorkExp` or another relevant experience-related column.


```python
df = pd.read_csv("survey-data.csv")

trend = (
    df.groupby("YearsCodePro")[["ConvertedCompYearly", "JobSatPoints_6"]]
    .median()
    .sort_index()
)

plt.figure(figsize=(10,5))

plt.plot(trend.index, trend["ConvertedCompYearly"], marker='o', label="Compensation")
plt.plot(trend.index, trend["JobSatPoints_6"], marker='o', label="Job Satisfaction")

plt.title("Trends in Compensation vs Job Satisfaction by Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Median Value")
plt.legend()
plt.xticks(rotation=45)
plt.grid(True, linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
```

##### 2.Line Chart of Job Satisfaction (`JobSatPoints_6`) Across Experience Levels

- Create a line chart to explore trends in job satisfaction (`JobSatPoints_6`) based on experience level.

- This chart will provide insight into how satisfaction correlates with experience over time


```python
%pip install pandas matplotlib -q

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("survey-data.csv")

df = df[['YearsCodePro', 'JobSatPoints_6']].dropna()

def clean_exp(x):
    if x == "Less than 1 year":
        return 0.5
    if x == "More than 50 years":
        return 50
    try:
        return float(x)
    except:
        return None

df['Experience'] = df['YearsCodePro'].apply(clean_exp)
df = df.dropna(subset=['Experience'])

exp_jobsat = df.groupby('Experience')['JobSatPoints_6'].median().sort_index()

plt.figure(figsize=(10, 5))
plt.plot(exp_jobsat.index, exp_jobsat.values, marker='o')

plt.title('Job Satisfaction Across Experience Levels')
plt.xlabel('Years of Professional Experience')
plt.ylabel('Median JobSatPoints_6')
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```
